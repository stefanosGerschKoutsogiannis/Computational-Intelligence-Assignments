# Clustering data analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_data_and_display_plots(data, imagename):
    groups = data.groupby("cluster_id")
    
    plt.figure(figsize=(10, 8))
    for cluster_id, cluster in groups:
        plt.scatter(cluster['x1'], cluster['x2'], label=f'Cluster {cluster_id}', alpha=0.6)
    
        plt.scatter(cluster['c_x1'].iloc[0], cluster['c_x2'].iloc[0], 
                    color='black', marker='x', s=200, label=f'Centroid {cluster_id}')

    plt.title('Scatter Plot of Clusters with Centroids', fontsize=16)
    plt.xlabel('x1', fontsize=14)
    plt.ylabel('x2', fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.savefig(f"output/clustering/plots/{imagename}")
    plt.show()
    
variance_data = pd.read_csv("output/clustering/clusters_variance.csv", header=None)
variance_data = variance_data.to_numpy()
```

```python
best_4_clusters = pd.read_csv("output/clustering/best_4_clusters.csv")
read_data_and_display_plots(best_4_clusters, "4_clusters.png")
print(f"Total variance with {variance_data[0][0]} clusters: {variance_data[0][1]}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_data(data, filename):
    correct_points = data[data.correct == 1]
    incorrect_points = data[data.correct == 0]
    
    plt.figure(figsize=(10,10))
    plt.scatter(correct_points.x1,correct_points.x2, marker="+")
    plt.scatter(incorrect_points.x1,incorrect_points.x2,marker="x")
    plt.title("Correctly and incorectly classified points", fontsize=16)
    plt.xlabel("x1", fontsize=14)
    plt.ylabel("x2", fontsize=14)
    plt.grid(True)
    plt.savefig(f"output/networks/plots/{filename}")
    plt.show()
```

```python
best_model2_data = pd.read_csv("output/networks/model2_predictions.csv", names=["x1","x2","correct"])

plot_data(best_model2_data, "accuracy_model2.png")
```

```python
best_model3_data = pd.read_csv("output/networks/model3_predictions.csv", names=["x1","x2","correct"])
plot_data(best_model3_data, "accuracy_model3.png")
```

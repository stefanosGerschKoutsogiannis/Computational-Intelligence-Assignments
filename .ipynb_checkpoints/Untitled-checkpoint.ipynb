```python
#generating data for clustering problem
x11 = np.random.uniform(-2, -1.6, 100)
x12 = np.random.uniform(1.6, 2, 100)
c1 = np.column_stack((x11, x12))
print(c1.shape)
x21 = np.random.uniform(-1.2, -0.8, 100)
x22 = np.random.uniform(1.6, 2, 100)
c2 = np.column_stack((x21, x22))
print(c2.shape)
x31 = np.random.uniform(-0.5, 0, 100)
x32 = np.random.uniform(1.6, 2, 100)
c3 = np.column_stack((x31, x32))
print(c3.shape)
x41 = np.random.uniform(-1.8, -1.4, 100)
x42 = np.random.uniform(0.8, 1.2, 100)
c4 = np.column_stack((x41, x42))
print(c4.shape)
x51 = np.random.uniform(-0.6, -0.2, 100)
x52 = np.random.uniform(0.8, 1.2, 100)
c5 = np.column_stack((x51, x52))
print(c5.shape)
x61 = np.random.uniform(-2, -1.6, 100)
x62 = np.random.uniform(0, 0.4, 100)
c6 = np.column_stack((x61, x62))
print(c6.shape)
x71 = np.random.uniform(-1.2, -0.8, 100)
x72 = np.random.uniform(0, 0.4 ,100)
c7 = np.column_stack((x71, x72))
print(c7.shape)
x81 = np.random.uniform(-0.4, 0, 100)
x82 = np.random.uniform(0.4, 0, 100)
c8 = np.column_stack((x81, x82))
print(c8.shape)
xo1 = np.random.uniform(-2, 0, 200)
xo2 = np.random.uniform(0, 2, 200)
co = np.column_stack((xo1, xo2))
print(co.shape)

data = np.concatenate((c1, c2, c3, c4, c5, c6, c7, c8, co), axis=0)
data.shape
```

```python
plt.scatter(data[:, 0], data[:, 1])
plt.show()
```

```python
dataset = pd.DataFrame(data)
dataset.to_csv('clustering_dataset.csv', header=False, index=False)
```

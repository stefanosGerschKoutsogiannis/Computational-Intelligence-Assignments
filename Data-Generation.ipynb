## Generation of datasets for the exercise

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_points = 8000
x1 = np.random.uniform(-1, 1, num_points)
x2 = np.random.uniform(-1, 1, num_points)

categories = []

for i in range(num_points):
    if (x1[i] - 0.5)**2 + (x2[i] - 0.5)**2 < 0.2 and x2[i] > 0.5:
        categories.append(1)
    elif (x1[i] - 0.5)**2 + (x2[i] - 0.5)**2 < 0.2 and x2[i] < 0.5:
        categories.append(2)
    elif (x1[i] + 0.5)**2 + (x2[i] + 0.5)**2 < 0.2 and x2[i] > -0.5:
        categories.append(1)
    elif (x1[i] + 0.5)**2 + (x2[i] + 0.5)**2 < 0.2 and x2[i] < -0.5:
        categories.append(2)
    elif (x1[i] - 0.5)**2 + (x2[i] + 0.5)**2 < 0.2 and x2[i] > -0.5:
        categories.append(1)
    elif (x1[i] - 0.5)**2 + (x2[i] + 0.5)**2 < 0.2 and x2[i] < -0.5:
        categories.append(2)
    elif (x1[i] + 0.5)**2 + (x2[i] - 0.5)**2 < 0.2 and x2[i] > 0.5:
        categories.append(1)
    elif (x1[i] + 0.5)**2 + (x2[i] - 0.5)**2 < 0.2 and x2[i] < 0.5:
        categories.append(2)
    elif x1[i] * x2[i] > 0:
        categories.append(3)
    else:
        categories.append(4)

data = pd.DataFrame({'x1': x1, 'x2': x2, 'category': categories})

train_data = data[:4000]
test_data = data[4000:]


print(train_data.head())
print(test_data.head())

train_data.to_csv("data/networks/classification_train.csv", header=False, index=False)
test_data.to_csv("data/networks/classification_test.csv",  header=False, index=False)
```

```python
colors = {1: 'red', 2: 'blue', 3: 'green', 4: 'purple'}

plt.figure(figsize=(7, 7))

for category, color in colors.items():
    subset = data[data['category'] == category]
    plt.scatter(subset['x1'], subset['x2'], s=10, color=color, label=category, alpha=0.7, marker="x")

plt.title("Dataset", fontsize=14)
plt.xlabel("x1", fontsize=10)
plt.ylabel("x2", fontsize=10)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend(title="Categories", fontsize=10)
plt.grid(alpha=0.3)
plt.show()
```
